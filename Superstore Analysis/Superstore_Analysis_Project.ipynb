# ***Data Visualization***
**Perform a correlation analysis between numerical columns**

```python
# Select only the numerical columns
numerical_columns = df.select_dtypes(include=['float64', 'int64'])
# Calculate the correlation matrix
correlation_matrix = numerical_columns.corr()
# Set up the matplotlib figure
plt.figure(figsize=(10, 8))
# Create a heatmap to visualize the correlation matrix
sns.heatmap(correlation_matrix, annot=True)
plt.title('Correlation Matrix of Numerical Columns')
# Display the plot
plt.show()
```

**Correlation:**
Correlation measures the relationship between two variables.
The correlation heatmap provided the relationships between different numerical columns in the dataset: 'Sales', 'Quantity', 'Discount', and 'Profit'.

Correlation values range from -1 to 1.

* 1 indicates a perfect positive correlation.
* -1 indicates a perfect negative correlation.
* 0 indicates no correlation.

**Insight**

**1. Sales and Profit (0.48):**

  There is a moderately strong positive correlation between 'Sales' and 'Profit'. This means that as sales increase, profit also tends to increase. This is a typical and expected relationship, as higher sales often lead to higher profits, assuming costs remain constant or grow at a slower rate.

**2. Sales and Quantity (0.20):** 

  There is a weak positive correlation between 'Sales' and 'Quantity'. This suggests that higher sales are somewhat associated with a larger quantity of items sold, but the relationship is not very strong. This could be due to variations in item prices affecting the total sales.

**3. Sales and Discount (-0.028):** 

  There is a very weak negative correlation between 'Sales' and 'Discount', indicating that discounts slightly reduce sales. This might seem counterintuitive but could be due to high discounts on low-priced items or ineffective discount strategies.

**4. Quantity and Profit (0.066):** 

  There is a very weak positive correlation between 'Quantity' and 'Profit'. This indicates that selling a higher quantity has a minor positive impact on profit. This weak correlation might be because of profit margins vary across products.

**5. Quantity and Discount (0.0087):**

  There is an almost negligible positive correlation between 'Quantity' and 'Discount', indicating that discounts have an insignificant impact on the quantity sold. This could mean that discounts are either not substantial enough to affect purchasing behavior or that they are applied to low-quantity items.

**6.  Discount and Profit (-0.22):**

  There is a weak negative correlation between 'Discount' and 'Profit'. This means that higher discounts are generally associated with lower profits, which is expected because discounts reduce the selling price and hence the profit margin.

**Actionable Steps:**

  * **Sales and Profit:** The strongest relationship is between sales and profit. This reinforces the importance of increasing sales to drive profit growth.

* **Discount Impact:** Discounts have a notable negative impact on profit and a very weak negative impact on sales, suggesting that discounts might not be the best strategy for driving sales and could harm profitability.

* **Quantity Relationships:** The weak correlations involving quantity suggest that other factors (like product pricing and cost structure) play a significant role in driving sales and profit.

***Generate a pie chart to visualize the proportion of sales  by region***

```python
#Check the 'Region' & 'Sales' column is present or not in the dataframe
if 'Region' in df.columns and 'Sales' in df.columns:
  top_region = df.groupby('Region')['Sales'].sum()  # Group the DataFrame by 'Region' and sum the 'Sales' for each region
  percentage=(top_region/top_region.sum())*100  # Calculate the percentage contribution of each region's sales to the total sales
  labels = ['East', 'West', 'Central', 'South'] # Define the labels for the pie chart
  sizes = [top_region['East'], top_region['West'], top_region['Central'], top_region['South']]  # Define the sizes of the pie chart slices based on the sales of each region
  colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']  # Define the colors for the pie chart slices
  explode = (0, 0.05, 0, 0) # Define the explosion parameters to slightly separate the 'West' slice from the rest
  plt.pie(sizes, explode=explode, labels=labels, colors=colors,autopct='%0.3f%%') # Create a pie chart with the defined parameters
  plt.axis('equal') # Ensure the pie chart is a circle
  plt.show()   # Display the pie chart
```

**scatter plot to analyze the relationship between sales and profit**

```python
#Check 'Sales' & 'Profit' columns are present or not in the dataframe
if 'Sales' in df.columns and 'Profit' in df.columns:
  plt.figure(figsize=(10,6))#Set matplotlib figure size
  sns.regplot(x='Sales',y='Profit',data=df,color='red')#Create a regression plot # Create a scatter plot with a regression line using seaborn's regplot
  plt.xlabel('Sales')#Set label of x-axis
  plt.ylabel('Profit')#Set label of y-axis
  plt.title('Scatter plot of Sales vs Profit')#Set title of the scatter pplot
  plt.grid(True)#enabel the grid
  plt.xticks(rotation=45)# Rotate the x-axis tick labels by 45
  plt.show()#Disply the plot
```

This scatter plot depicts the relationship between sales and profit.
1. **Positive Correlation**: There is a visible positive trend in the scatter plot, as indicated by the upward-sloping red line. This suggests that, generally, higher sales are associated with higher profits.

2. **Variance**: The data points are widely spread, especially for higher sales values. This indicates variability in profit for given sales amounts. Some points lie far from the trend line, indicating outliers or potential anomalies.

3. **Outliers**: There are several points that are significantly distant from the trend line. For example, there are cases with high sales but much lower profit (or even losses) and vice versa. These outliers can be important for deeper analysis to understand the causes.

4. **Range of Data**: The range of sales varies from nearly 0 to over 20,000, while profit ranges from approximately -6,000 to 8,000. The wide range suggests a diverse dataset with potentially different categories or types of sales transactions.

5. **Concentration**: A large number of points are concentrated at the lower end of the sales spectrum. This may indicate that most transactions have smaller sales values with a corresponding cluster of profit values around the trend line.

6. **Regression Line**: The red regression line with a shaded confidence interval gives an idea of the central tendency and the reliability of the trend. The confidence interval widens with higher sales, indicating more uncertainty or variability in profit prediction at higher sales levels.
